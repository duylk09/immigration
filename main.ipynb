Connect to file statistics.xlsx <br>
Get the data from sheet Data_Occ_D01


```python
import pandas as pd

file_path = 'statistics.xlsx'
sheetname = 'Data_Occ_D01'

df = pd.read_excel(file_path,sheetname,skiprows=1)

```

Create list of Nationalities from the sheet

```python
import math

# Get the data of nationality column then remove all duplicates
column_nationality_data = df['Nationality']
column_nationality_list = list(set(column_nationality_data.tolist()))

# there is a 'nan' value, remove this value
filtered_countries = [country for country in column_nationality_list if isinstance(country, str) or (isinstance(country, float) and not math.isnan(country))]

#sorted the list
print(sorted(filtered_countries))
```

Display a graph to show the trend of the number of applications from the input of Nationality

```python
import matplotlib.pyplot as plt

# set up some fixed input

fixed_Vstype = 'Worker'
fixed_Vssubtype = ['Skilled Worker','Skilled Worker - Health & Care']

fixed_df = df [(df['Visa type'] == fixed_Vstype) & (df['Visa type subgroup'].isin(fixed_Vssubtype))]
# input nationality
filter_nationality = 'Vietnam'

selected_df = fixed_df[fixed_df['Nationality'] == filter_nationality]

# group the data by year and sum the Applicaions
yearly_data = selected_df.groupby('Year')['Applications'].sum().reset_index()

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(yearly_data['Year'], yearly_data['Applications'], marker='o')

# Customize the plot
plt.title('Trend of Applications Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Applications')
plt.grid(True)

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
